### Remover outliers

```python

def remove_outliers_zscore(df, column, threshold=3):
    mean = df[column].mean()
    std = df[column].std()
    z_scores = (df[column] - mean) / std
    df_clean = df[abs(z_scores) < threshold]
    rows_removed = df.shape[0] - df_clean.shape[0]
    return df_clean, rows_removed



for column in ['price','host_listings_count','accommodates','extra_people',
 'bathrooms','bedrooms','beds','minimum_nights']:
    features_df, linhas_removidas = remove_outliers_zscore(features_df,column)
    print(f'Foram removidas {linhas_removidas} ')
```
